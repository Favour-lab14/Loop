```python
student_info = {"student_name":"James", "age":20, "course":"Python"}

print(student_info)
```

```python
type(student_info)
```

```python
#accessing elements in a dictionary

student_info["student_name"]
```

```python
student_info["age"]
```

```python
student_info["course"]
```

```python
#access the keys in a dictionary

student_info.keys()
```

```python
#access the values in a dictionary

student_info.values()
```

```python
#access the items in a dictionary

student_info.items()
```

```python
student_info2 = {"student_name":"James", "age":20, "course":["python", "SQL"]}

print(student_info2)
```

```python
course_list = student_info2["course"]
```

```python
print(course_list)
```

```python
course_list[0]
```

```python
#modify items in a dictionary

student_info2["age"] = 22
```

```python
print(student_info2)
```

```python
#add new item

student_info2["location"] = "Nigeria"
```

```python
print(student_info2)
```

```python
#delete an item in a dictionary

del student_info2["age"]
```

### Nested Dictionaries

```python
students_info = {"student_1": {"student_name":"James", "age":20, "course":"Python"},
                 "student_2": {"student_name":"Alice", "age":22, "course":"SQL"}}

print(students_info)
```

```python
students_info["student_1"]
```

```python
students_info["student_1"]["student_name"]
```

```python
students_info["student_2"]["course"]
```
